```python
import pandas as pd 
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
```

```python
df
```

```python
missing_value=df.isnull().sum()
print(missing_value)
```

```python
#print(df['completed'].unique())
print(df['Sleep Disorder'].unique())
df.fillna("No Sleep Disorder", inplace=True)
```

```python
missing_value=df.isnull().sum().sum()
print(missing_value)
print(df['Sleep Disorder'].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Plotting scatterplot with different colors for gender
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Quality of Sleep', hue='Gender', palette='coolwarm')
plt.xlabel('Age')
plt.ylabel('Quality of Sleep')
plt.title('Scatter Plot: Age vs Quality of Sleep (Colored by Gender)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Quality of Sleep', y='Stress Level', palette='cool')
plt.xlabel('Quality of Sleep')
plt.ylabel('Stress Level')
plt.title('Scatter Plot: Age vs Quality of Sleep (Colored by Gender)')
plt.show()
```

```python

plt.figure(figsize=(10, 6))

# Create a count plot to visualize Age vs. BMI Category
sns.countplot(data=df, x='Age', hue='BMI Category')

# Customize the plot
plt.title('Age vs. BMI Category')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='BMI Category')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Physical Activity Level', y='Heart Rate', hue='Gender', palette='coolwarm')
plt.xlabel('Physical Activity Level')
plt.ylabel('Heart Rate')
plt.title('Scatter Plot: ')
plt.show()
```

```python
# Visualize the relationship between 'Quality of Sleep' and 'Sleep Disorder' using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Quality of Sleep', hue='Sleep Disorder', palette='coolwarm', multiple='stack')
plt.xlabel('Quality of Sleep')
plt.ylabel('Count')
plt.title('Histogram: Quality of Sleep vs. Sleep Disorder')
plt.show()
```
