```python
import pandas as pd 
df=pd.read_csv('movies.csv')
df
```

```python
df.isnull().sum()
print(df['GENRE'].unique())
```

```python
df.drop('Gross',axis=1,inplace=True)
```

```python
# Replace missing values in 'GENRE' with 'Unknown' using fillna() with a dictionary
df = df.fillna({'GENRE': 'Unknown'})

df
df.isnull().sum()
```

```python
df['RATING'] = df['RATING'].fillna(df['RATING'].mean())

df.isnull().sum()
```

```python
# Delete rows where 'YEAR' column has None value
df = df.dropna(subset=['YEAR'])
df.isnull().sum()
```

```python
print(df[df['RunTime'].isnull()])
df.shape
```

```python
# Delete rows where 'RunTime' column has None value
df = df.dropna(subset=['RunTime'])
df.isnull().sum()
```

```python
# Delete rows where 'VOTES' column has None value
df = df.dropna(subset=['VOTES'])
df.isnull().sum()
```

```python
unique_values = df['ONE-LINE'].unique()
print(unique_values)
```

```python
df['GENRE'] = df['GENRE'].str.strip()  # Remove leading and trailing whitespace
df['GENRE'] = df['GENRE'].str.split(',').apply(lambda x: [genre.strip() for genre in x])  # Split genres and strip whitespace

print(df['GENRE'])
```

```python
df['ONE-LINE'] = df['ONE-LINE'].str.strip()  # Remove leading and trailing whitespace

# Display the cleaned 'ONE-LINE' column
print(df['ONE-LINE'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of numerical column
plt.hist(df['RunTime'], bins=20)
plt.xlabel('RunTime')
plt.ylabel('Frequency')
plt.title('Distribution of RunTime')
plt.show()

```

```python
import pandas as pd

# Assuming df is your DataFrame containing the data
# Replace 'df' with the name of your DataFrame

# Check the data type of the 'VOTES' column
print(df['VOTES'].dtype)

# Convert 'VOTES' column to string type (object dtype) if it's not already
if not pd.api.types.is_string_dtype(df['VOTES']):
    df['VOTES'] = df['VOTES'].astype(str)

# Clean 'VOTES' column to remove commas and convert to numeric type
df['VOTES'] = pd.to_numeric(df['VOTES'].str.replace(',', ''), errors='coerce')

# Print updated data types
print(df.dtypes)
```
